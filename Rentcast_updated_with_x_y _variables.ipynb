```python

import json
import requests
from dotenv import load_dotenv, find_dotenv
import os

load_dotenv(find_dotenv("local_keys.env"))


```

```python
key = os.getenv("rentcast")
```

```python
type(key)
```

```python
state = input(str("What state would you like to view?: ")).capitalize()
```

```python
zipcode = input("What zip code would you like to search for?: ")
```

```python
radius = input("How far would you like to view around that zip code in mile?: ")
```

```python
url = f"https://api.rentcast.io/v1/avm/value?address={state}%2C%20{zipcode}&propertyType=Single%20Family&maxRadius={radius}&compCount=25"

headers = {
    "accept": "application/json",
    "X-Api-Key": key
}

response = requests.get(url, headers=headers)
responsedata = response.json()

#listings = pd.json_normalize(responsedata)
```

```python
import pandas as pd
print(responsedata)
df = pd.DataFrame(responsedata)
```

```python
records = []
for key, value in responsedata.items():
    record = {}
    if key == 'comparables':
        print(value)
        for comparable in value:
            for ix, v in comparable.items():
                record[ix] = v
            # record[] = comparable

    record[key] = value
    records.append(record)
pd.DataFrame(records)
```

```python
houses = []
for record in responsedata['comparables']:
    print(record)
    houses.append(record)
```
