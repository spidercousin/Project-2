```python
from sklearn.datasets import make_regression
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
import pandas as pd

file = pd.read_csv('dropped_cleaned.csv')
df = pd.DataFrame(file)
```

```python
df.head()
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
scaler = StandardScaler(with_std=False)
X_train_transformed = scaler.fit_transform(X_train)
y_train_transformed = scaler.fit_transform(y_train.values.reshape(-1, 1))
```

```python
model = Ridge(alpha=1)
model.fit(X_train_transformed, y_train_transformed)
```

```python
scaler_X = StandardScaler(with_std=False)
scaler_y = StandardScaler(with_std=False)

X_train_transformed = scaler_X.fit_transform(X_train)
y_train_transformed = scaler_y.fit_transform(y_train.values.reshape(-1, 1))

X_test_transformed = scaler_X.transform(X_test)
y_test_transformed = scaler_y.transform(y_test.values.reshape(-1, 1))
y_predicted = model.predict(X_test_transformed)
```

```python
mean_squared_error(y_test, y_predicted)
```

```python
from sklearn.linear_model import RidgeCV
model_cv = RidgeCV(alphas=[0.001, 0.01, 0.1, 1, 10])
model_cv = model_cv.fit(X_train_transformed, y_train)
```

```python
model_cv.alpha_
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(X_train_transformed, y_train)
y_predicted_lr = lr_model.predict(X_test_transformed)
mean_squared_error(y_test, y_predicted_lr)
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_model = Lasso(alpha=1)
lasso_model.fit(X_train_transformed, y_train)
```

```python
lasso_model.coef_
```

```python
y_predicted_lasso = lasso_model.predict(X_test_transformed)
```

```python
mean_squared_error(y_test, y_predicted_lasso)
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error, r2_score

model5 = ElasticNet(alpha=0.1, l1_ratio=0.5)
model5.fit(X_train_transformed, y_train_transformed)

y_predicted_elasticnet = model5.predict(X_test_transformed)
mse = mean_squared_error(y_test, y_predicted_elasticnet)
r2 = r2_score(y_test, y_predicted_elasticnet)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
